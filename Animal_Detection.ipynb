```python
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input, decode_predictions
# We called the necessary libraries.
```

```python
import numpy as np
# "Numpy" kütüphanesini çağırdık
```

```python
model = VGG16(weights = "imagenet" ) # "VGG16" is a neural network mimairis. We created the variable "model" and defined the weight as "imagenet".
```

```python
img_path = "cat.png" # We define an animal image that we want it to recognize and pre-upload to a variable called" img_path".
img = image.load_img(img_path, target_size=(224,224)) # We're turning it into the necessary dimension.
```

```python
x = image.img_to_array(img)
x = np.expand_dims(x , axis=0)
x = preprocess_input(x)
# We've done the necessary procedures.
```

```python
preds = model.predict(x) # We have defined the variable to which we will throw the image to try the model.
```

```python
preds
```

```python
print(decode_predictions(preds, top=5)) # We typed "decode_predictions(preds, top=5)" into the print command to show which animal it was closer to.
# We got the top 5 values with" top=5".
```
